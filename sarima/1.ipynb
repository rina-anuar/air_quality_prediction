```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf
from statsmodels.tsa.stattools import adfuller
import seaborn as sns
```

```python
df = pd.read_csv('Book1.csv')
df
```

```python
df['Date'] = pd.date_range(start='2017-03-22', periods= len(df), freq='D')
```

```python
# 2
df.set_index('Date', inplace=True)

# 3
df_clean = df.dropna(subset=['1'])

# 4
df_resampled = df_clean['1'].resample('ME').mean()
```

```python
sns.set(style="whitegrid")

plt.figure(figsize=(12, 6))
# Plot the resampled data
sns.lineplot(data=df_resampled, color='blue', marker='o')

plt.xlabel('Date')
plt.ylabel('Average Value Of AQI')
plt.show()
```
